# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
import selenium
import pandas as pd
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(r'C:\Windows\chromedriver.exe')
driver
```

```python
url ='https://www.naukri.com/'
driver.get(url)
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input")
designation.send_keys("Data Analyst")
```

```python
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
skill_required=[]
```

```python
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    tital=i.text
    job_title.append(tital)
    
    
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
    
    
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
    
    
skill_tags=driver.find_elements(By.XPATH, '//div[@class="ellipsis job-description"]')
for i in skill_tags[0:10]:
    skill=i.text
    skill_required.append(skill)
    
    
```

```python
print(len(job_title),len(job_location),len(company_name),len(skill_required))
```

```python
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Companies':company_name,'Skill':skill_required})
df = df.apply(lambda x: x.str.replace('\n', ' ,'))
df
```

# Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpage https://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make the DataFrame.

```python
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

# Connecting WebDriver
driver = webdriver.Chrome(r'C:\Windows\chromedriver.exe')

# Opening azquotes.com
url = 'https://www.jagranjosh.com/'
driver.get(url)
# Entering "gk" in the search input field

button = driver.find_element_by_link_text('GK')
button.click()
```

```python
# Entering "List of all Prime Ministers of India" in the search input field
button = driver.find_element_by_xpath('//*[@id="popluarGK"]/ul/li[2]/a')
button.click()
```

```python
list_of_PM = []
```

```python
# Scrape the data
table_rows = driver.find_elements_by_xpath('//div[@class="table-box"]/table/tbody/tr')
for row in table_rows[1:]:
    columns = row.find_elements_by_tag_name('td')
    if len(columns) >= 4:
        name = columns[0].text
        born_dead = columns[1].text
        term_of_office = columns[2].text
        remarks = columns[3].text
        list_of_PM.append([name, born_dead, term_of_office, remarks])
    else:
        print("Skipping row with insufficient columns:", [column.text for column in columns])
```

```python
df = pd.DataFrame(list_of_PM, columns=['Name', 'Born-Dead', 'Term of Office', 'Remarks'])
df = df.apply(lambda x: x.str.replace('\n', ' ,'))
df
```

# Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.

3. Then click on 50 most expensive cars in the world..

```python
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException

# Connecting WebDriver
driver = webdriver.Chrome(r'C:\Windows\chromedriver.exe')

# Opening azquotes.com
url = 'https://www.motor1.com/'
driver.get(url)
```

```python
button = driver.find_element_by_xpath('//*[@id="page_index_index_index"]/div[2]/div/div/div[1]/div')
button.click()
```

```python
# Entering "List of 50 most expensive cars in the world" in the search input field
button = driver.find_element_by_xpath('/html/body/div[4]/div[1]/div[3]/ul/li[11]/a')
button.click()
```

```python
# Scraping the data
car_list = []
table_rows = driver.find_elements_by_xpath('//*[@id="page_index_submodels_ratings-list"]/div[7]')
for row in table_rows:
    name = row.find_element_by_xpath('//div[@class="model-title"]').text
    model = row.find_element_by_xpath('//div[@class="model-subtitle"]').text
    body = row.find_element_by_xpath('//div[@class="model-type"]').text
    year = row.find_element_by_xpath('//div[@class="model-year"]').text
    rating = row.find_element_by_xpath('//div[@class="model-verdict"]').text
    
    car_list.append([name, model, body, year, rating])

# Creating a DataFrame
df = pd.DataFrame(car_list, columns=['Car Name', 'Model', 'Body', 'Year', 'Rating'])
df = df.apply(lambda x: x.str.replace('\n', ', '))
df.head(50)  # Displaying the first 50 most expensive cars
```
